```python
import pandas as pd
import numpy as np
laliga_df=pd.read_csv("Laliga.csv")
laliga_df.head()
```

```python
laliga_df.columns=laliga_df.iloc[0]   # Since all the column names are Unnamed, so making the 1st row as the column names
laliga_df.drop([0],inplace=True)  #Dropping the 1st row. Or else 1st row and column names will be similar.
laliga_df
```

# 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. 

```python
for each in laliga_df.columns:
        laliga_df[each].replace('-',0,inplace=True)

laliga_df #all the '-' are replaced with 0
```

# 2.Print all the teams which have started playing between 1930-1980. 

```python
import re

# In the Debut column there are some strings like 1945-46. I have split those strings, and have taken only the 1st value.
#For eg: 1945-46 is splitted to 1945 and 46. I have taken only 1945.
for each in laliga_df['Debut']:
    if re.match("[0-9]{4}-[0-9]{2}",each):
        laliga_df['Debut'].replace(each,each.split('-')[0],inplace=True)


#The type of each elements in the column is string. So I have converted all the values to int, for convenient calculation
laliga_df['Debut']=pd.to_numeric(laliga_df['Debut'], errors='coerce')
```

```python
count =0
for each in laliga_df['Debut']:
    if (each>=1930 and each<=1980):
        count=count+1

print("Total number of teams that played between 1930 and 1980 are:",count)
```

# 3. Print the list of teams which came Top 5 in terms of points

```python
laliga_df['Points']=pd.to_numeric(laliga_df['Points'], errors='coerce') #The data type of Points is String. So I have converted to integer

laliga_df.sort_values(by='Points',ascending=False)[['Team','Points']].iloc[0:5,:]
```

# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
# Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(Team):
    return(Team['GoalsFor'] - Team['GoalsAgainst'])

laliga_df['GoalsFor']=pd.to_numeric(laliga_df['GoalsFor'], errors='coerce')
laliga_df['GoalsAgainst']=pd.to_numeric(laliga_df['GoalsAgainst'], errors='coerce')

laliga_df['Goal_diff_count']=laliga_df.apply(Goal_diff_count,axis=1)

laliga_df[['Team','Goal_diff_count']] #Prints the Goal_Diff_Count and the corresponding Teams
```

```python
max_index=laliga_df['Goal_diff_count'].idxmax() #Finds the index of the maximum goal difference count and the team
laliga_df[['Team','Goal_diff_count']].loc[max_index]
```

```python
min_index=laliga_df['Goal_diff_count'].idxmin() #Finds the index of the minimum goal difference count and the team
laliga_df[['Team','Goal_diff_count']].loc[min_index]
```

# 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100
 If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
def Winning_Percent(df):
    try:
        return((df['GamesWon']/df['GamesPlayed'])*100)
    except:
        return(0)      #This will return zero if 'division by zero error occurs'

laliga_df['GamesWon']=pd.to_numeric(laliga_df['GamesWon'], errors='coerce')
laliga_df['GamesPlayed']=pd.to_numeric(laliga_df['GamesPlayed'], errors='coerce')


laliga_df['Winning Percent']=laliga_df.apply(Winning_Percent,axis=1)
laliga_df['Winning Percent']
```

```python
laliga_df.nlargest(5,'Winning Percent')
```

# 6.Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
Eg: Best Position                Points

        1                              25000

        2                              7000      

```python
# Both BestPosition and Points are converted to numeric to ease arithmetic operations
laliga_df['BestPosition']=pd.to_numeric(laliga_df['BestPosition'], errors='coerce')
laliga_df['Points']=pd.to_numeric(laliga_df['Points'], errors='coerce')
group_by_BestPosition=laliga_df[['Team','BestPosition','Points']].groupby('BestPosition') #Grouped according to BestPosition
group_by_BestPosition.get_group(1) #This will print the values for Best Position 1
```

```python
group_by_BestPosition.sum() # This shows the sum of their points for all positions
```
